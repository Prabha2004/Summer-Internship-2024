```python
import matplotlib.pyplot as plt


team_name = input("Enter the team name: ")


team_data = data[data['Team'] == team_name]


plt.figure(figsize=(10, 6))
plt.bar(team_data['Opponent'], team_data['Pass Accuracy %'], color='skyblue')
plt.xlabel('Opponent')
plt.ylabel('Pass Accuracy %')
plt.title(f'Pass Accuracy % of {team_name} against different opponents')


plt.show()

```

```python
avg_freekicks=data.groupby('Team')['Free Kicks'].mean()
print(avg_freekicks)
```

```python
plt.figure(figsize=(10,6))
avg_freekicks.plot(kind='bar',color='Red')
plt.title("Average Free Kicks")
plt.xlabel("Countries")
plt.ylabel("Free Kicks")
plt.show()
```

```python
import pandas as pd


non_numeric_cols = data.select_dtypes(exclude=["number"]).columns


print(f"Non-numeric columns: {non_numeric_cols}")


data.drop(non_numeric_cols, axis=1, inplace=True)


correlation_matrix = data.corr()


for col in non_numeric_cols:
    try:
        data[col] = pd.to_numeric(data[col])
    except:

        pass


correlation_matrix = data.corr()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
